Exercise 5.1 ให้เขียน function ชื่อ day_of_year(day, month ,year) โดยมีการคืนค่า คือ day_of_years เป็นวันที่ลำดับที่เท่าใดของปีคริสตศักราช year
ปีที่เป็น Leap Year เดือนกุมภาพันธ์จะมี 29 วัน
ให้สร้างฟังก์ชัน is_leap เพื่อตรวจสอบ leap year แยกออกมา และให้ฟังก์ชัน day_of_year เรียกใช้ is_leap อีกที


````python
days_in_month = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]

def is_leap(year):
  if year%4 == 0 and year%100 !=0 or year%400 == 0:
    return True
  else :
    return False


def day_of_year(day, month, year):
  sumday = 0
  for i in range(month):
    sumday = sum(days_in_month[:month-1])
    sumday = sumday+day
  if is_leap(year) and month > 2 :
    sumday = sumday + 1
  return sumday





# Testcases
print("Normal year (2023):")
print(f"1 Jan 2023: {day_of_year(1, 1, 2023)}")  # ควรได้ 1
print(f"1 Feb 2023: {day_of_year(1, 2, 2023)}")  # ควรได้ 32
print(f"1 Mar 2023: {day_of_year(1, 3, 2023)}")  # ควรได้ 60
print(f"31 Dec 2023: {day_of_year(31, 12, 2023)}")  # ควรได้ 365

print("\nLeap year (2024):")
print(f"1 Jan 2024: {day_of_year(1, 1, 2024)}")  # ควรได้ 1
print(f"29 Feb 2024: {day_of_year(29, 2, 2024)}")  # ควรได้ 60
print(f"1 Mar 2024: {day_of_year(1, 3, 2024)}")  # ควรได้ 61
print(f"31 Dec 2024: {day_of_year(31, 12, 2024)}")  # ควรได้ 366

print("\nCentury year, not a leap year (1900):")
print(f"1 Mar 1900: {day_of_year(1, 3, 1900)}")  # ควรได้ 60

print("\nCentury year, leap year (2000):")
print(f"1 Mar 2000: {day_of_year(1, 3, 2000)}")  # ควรได้ 61

print("\nMid-year dates:")
print(f"15 Jun 2023: {day_of_year(15, 6, 2023)}")  # ควรได้ 166
print(f"15 Jun 2024: {day_of_year(15, 6, 2024)}")  # ควรได้ 167 (Leap year)

print("\nLast day of each quarter:")
print(f"31 Mar 2023: {day_of_year(31, 3, 2023)}")  # ควรได้ 90
print(f"30 Jun 2023: {day_of_year(30, 6, 2023)}")  # ควรได้ 181
print(f"30 Sep 2023: {day_of_year(30, 9, 2023)}")  # ควรได้ 273
print(f"31 Dec 2023: {day_of_year(31, 12, 2023)}")  # ควรได้ 365
````

Exercise 5.2 ให้เขียนฟังก์ชัน เพิ่มเติมเป็น date_diff
รับข้อมูลในรูปแบบ “dd-mm-yyyy” เช่น date_diff(“1-1-2018”, “1-1-2020”) จะได้ 731 วัน date_diff(“25-12-1999”, “9-3-2000”) จะได้ 76 วัน
ให้เขียนฟังก์ชัน day_in_year โดยจะส่งค่าจำนวนวันของปี (365 หรือ 366) โดยรับข้อมูลเป็น ปี
ส่งคืนข้อมูลเป็นจำนวนวันตั้งแต่วันที่แรก จนถึงวันที่สอง โดยรวมทั้ง 2 วันนั้นเข้าไปด้วย
ให้สมมติว่าวันแรก จะต้องมาก่อนวันที่สองเสมอ ดังนั้นไม่ต้องตรวจสอบ


````python
days_in_month = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]

def is_leap(year):
  if year%4 == 0 and year%100 !=0 or year%400 == 0:
    return True
  else :
    return False

def day_of_year(day, month, year):
  sumday = 0
  for i in range(month):
    sumday = sum(days_in_month[:month-1])
    sumday = sumday+day
  if is_leap(year) and month > 2 :
    sumday = sumday + 1
  return sumday

def day_in_year(year):
    return 366 if is_leap(year) else 365

def date_diff(date1, date2):
    allsumday = 0
    day1,month1,year1 = date1.split('-')
    day2,month2,year2 = date2.split('-')
    if int(year1) == int(year2) :
      daysum = day_in_year(int(year1))
      day1pass = day_of_year(int(day1), int(month1), int(year1))
      day2pass = day_of_year(int(day2), int(month2), int(year2))
      result = daysum - (daysum - day2pass)
      return result
    else :
      daysum = day_in_year(int(year1)) + day_in_year(int(year2))
      for i in range(int(year1)+1,int(year2)):
        if is_leap(int(year1)+1):
          daysum = daysum + 366
        else:
          daysum = daysum + 365
      day1pass = day_of_year(int(day1), int(month1), int(year1))
      day2pass = day_of_year(int(day2), int(month2), int(year2))
      result = daysum - day1pass - (day_in_year(int(year2)) - day2pass)
      return result+1


# ทดสอบฟังก์ชัน
print(date_diff("1-1-2018", "1-1-2020"))  # ควรได้ 731
print(date_diff("25-12-1999", "9-3-2000"))  # ควรได้ 76
print(date_diff("1-1-2000", "1-1-2001"))  # ควรได้ 367 (ปี 2000 เป็น leap year)
print(date_diff("1-1-2001", "1-1-2002"))  # ควรได้ 366 (ปี 2001 ไม่ใช่ leap year)
print(date_diff("1-3-2000", "1-3-2001"))  # ควรได้ 367 (ข้ามปี leap year)
print(date_diff("1-1-2023", "31-12-2023"))  # ควรได้ 365 (ทั้งปี 2023)
````

Exercise 5.3 จากโปรแกรม 5.2 ให้เขียนฟังก์ชัน date_diff  เพิ่มเติม โดยให้มีการตรวจสอบ
วันที่ต้องเป็นวันที่ถูกต้องของเดือนนั้นๆ
เดือนต้องอยู่ระหว่าง 1-12
เดือนกุมภาพันธ์ของปีที่มี Leap Year เท่านั้นที่จะมี 29 วันได้
หากข้อมูล Input ผิดพลาด ให้ Return -1


````python
days_in_month = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]

def is_leap(year):
    return year % 4 == 0 and (year % 100 != 0 or year % 400 == 0)

def day_of_year(day, month, year):
    sumday = sum(days_in_month[:month-1]) + day
    if is_leap(year) and month > 2:
        sumday += 1
    return sumday

def day_in_year(year):
    return 366 if is_leap(year) else 365

def is_valid_date(day, month, year):
    if month < 1 or month > 12:
        return False
    if day < 1 or day > days_in_month[month-1]:
        return False
    if month == 2 and is_leap(year) and day > 29:
        return False
    if month == 2 and not is_leap(year) and day > 28:
        return False
    return True

def date_diff(date1, date2):
    try:
        day1, month1, year1 = map(int, date1.split('-'))
        day2, month2, year2 = map(int, date2.split('-'))
    except ValueError:
        return -1

    if not is_valid_date(day1, month1, year1) or not is_valid_date(day2, month2, year2):
        return -1

    if year1 == year2:
        day1pass = day_of_year(day1, month1, year1)
        day2pass = day_of_year(day2, month2, year2)
        return day2pass - day1pass
    else:
        if year1 > year2 or (year1 == year2 and day_of_year(day1, month1, year1) > day_of_year(day2, month2, year2)):
            day1, month1, year1, day2, month2, year2 = day2, month2, year2, day1, month1, year1

        daysum = day_in_year(year1) - day_of_year(day1, month1, year1) + 1
        daysum += day_of_year(day2, month2, year2)

        for year in range(year1 + 1, year2):
            daysum += day_in_year(year)

        return daysum

# ทดสอบฟังก์ชัน
print(date_diff("1-1-2018", "1-1-2020"))  # ควรได้ 730
print(date_diff("25-12-1999", "9-3-2000"))  # ควรได้ 75
print(date_diff("1-1-2000", "1-1-2001"))  # ควรได้ 366 (ปี 2000 เป็น leap year)
print(date_diff("1-1-2001", "1-1-2002"))  # ควรได้ 365 (ปี 2001 ไม่ใช่ leap year)
print(date_diff("1-3-2000", "1-3-2001"))  # ควรได้ 366 (ข้ามปี leap year)
print(date_diff("1-1-2023", "31-12-2023"))  # ควรได้ 364 (ทั้งปี 2023)

# ทดสอบกรณีข้อมูลไม่ถูกต้อง
print()
print(date_diff("31-4-2023", "1-5-2023"))  # ควรได้ -1 (เดือนเมษายนมี 30 วัน)
print(date_diff("29-2-2023", "1-3-2023"))  # ควรได้ -1 (ปี 2023 ไม่ใช่ leap year)
print(date_diff("1-13-2023", "1-1-2024"))  # ควรได้ -1 (เดือนไม่ถูกต้อง)
print(date_diff("abc", "1-1-2024"))  # ควรได้ -1 (รูปแบบวันที่ไม่ถูกต้อง)
````

- 0 เมื่อ string ทั้ง 2 ตัว เหมือนกัน โดยไม่สนใจ case
- 1 เมื่อ string 1 จะเหมือน string 2 โดยการลบและ/หรือเพิ่มตัวอักษรเพียง 1 ตัว

````python
# str1 = 'Dog'
# str2 = 'dog'

def Convert(str1, str2):
    str1 = list(str1.lower())
    str2 = list(str2.lower())
    return str1, str2

def find_mismatch(str1, str2):
    x = 0
    str1, str2 = Convert(str1, str2)
    i = 0
    while i < len(str1):
        j = 0
        while j < len(str2):
            # print(f'{str1} - {str2}')
            # print(f'{str1[i]} - {str2[j]}')
            if str1[i] == str2[j]:
                str1.pop(i)
                str2.pop(j)
                if i > 0:
                    i -= 1
                break
            else:
                j += 1
        else:
            i += 1
    if len(str1)>len(str2):
      x =  len(str1) - len(str2)
    elif len(str1)<len(str2):
      x =  len(str2) - len(str1)
    else :
      if len(str1) == 0 and len(str2) == 0 :
        return 0#,str1,str2,x

    if x >= 0 and x < 2:
      return 1#,str1,str2,x
    else :
      return 2#,str1,str2,x



print(find_mismatch('Python', 'Java'))
print(find_mismatch('book', 'boot'))
print(find_mismatch('fan', 'fun'))
print(find_mismatch('sin', 'sink'))
print(find_mismatch('Dog', 'dog'))
print(find_mismatch('spoke', 'poke'))
print(find_mismatch('poker', 'poke'))

````

4. ให้เขียนฟังก์ชัน find_word_horizontal ซึ่งจะรับ list ขนาด 2 มิติ ของตัวอักษร (คล้ายกับเกม cross word) และ word ซึ่งเป็น string เป็น argument โดยฟังก์ชันนี้ จะค้นหาตามแถวของ list เพื่อหาตำแหน่งที่มีอักษรตรงกับคำใน word ถ้าพบ จะส่งคืน row index และ column index ของตัวอักษรแรก ถ้าไม่พบจะส่งคืน None

```
crosswords=[['s','d','o','g'],
            ['c','u','c','m'],
            ['a','c','a','t'],
            ['t',' e','t','k']]
	word='cat'
	find_word_horizontal(crosswords,word)
```

Output
   [2,1]


````python
def find_word_horizontal(crossword, word):
    row_index = 0
    for row in crossword:
        row_str = ''.join(row)
        col_index = row_str.find(word)
        if col_index != -1:
            return (row_index, col_index)
        row_index += 1
    return None

crossword = [['s','d','o','g'],
            ['c','u','c','m'],
            ['a','c','a','t'],
            ['t',' e','t','k']]
word='cat'
find_word_horizontal(crossword, word)
````

5. ให้เขียนฟังก์ชัน find_word_vertical ซึ่งจะรับ list ขนาด 2 มิติ ของตัวอักษร (คล้ายกับเกม cross word) และ word ซึ่งเป็น string เป็น argument โดยฟังก์ชันนี้ จะค้นหาตามคอลัมน์ของ list เพื่อหาตำแหน่งที่มีอักษรตรงกับคำใน word ถ้าพบ จะส่งคืน row index และ column index ของตัวอักษรแรก ถ้าไม่พบจะส่งคืน None

ตัวอย่าง
```
	crosswords=[['s','d','o','g'],
              ['c','u','c','m'],
              ['a','c','a','t'],
              ['t',' e','t','k']]
	word='cat'
	find_word_vertical(crosswords,word)
```

Output
   [1,0]


````python
def find_word_vertical(crossword, word):
    num_rows = len(crossword)
    num_cols = len(crossword[0]) if num_rows > 0 else 0
    for col_index in range(num_cols):
        col_str = ''.join(crossword[row_index][col_index] for row_index in range(num_rows))
        row_index = col_str.find(word)
        if row_index != -1:
            return (row_index, col_index)
    return None
crossword = [['s','d','o','g'],
            ['c','u','c','m'],
            ['a','c','a','t'],
            ['t',' e','t','k']]
word='cat'
print(find_word_vertical(crossword, word))
````

````python
def find_word_horizontal(crossword, word):
    row_index = 0
    for row in crossword:
        row_str = ''.join(row)
        col_index = row_str.find(word)
        if col_index != -1:
            return (row_index, col_index)
        row_index += 1
    return None

def find_word_vertical(crossword, word):
    num_rows = len(crossword)
    num_cols = len(crossword[0]) if num_rows > 0 else 0
    for col_index in range(num_cols):
        col_str = ''.join(crossword[row_index][col_index] for row_index in range(num_rows))
        row_index = col_str.find(word)
        if row_index != -1:
            return (row_index, col_index)
    return None

def capitalize_word_in_crossword(crossword, word):
    def capitalize_in_matrix(start_row, start_col, direction):
        if direction == 'horizontal':
            for i in range(len(word)):
                crossword[start_row][start_col + i] = crossword[start_row][start_col + i].upper()
        elif direction == 'vertical':
            for i in range(len(word)):
                crossword[start_row + i][start_col] = crossword[start_row + i][start_col].upper()
    result = find_word_horizontal(crossword, word)
    if result:
        row_index, col_index = result
        capitalize_in_matrix(row_index, col_index, 'horizontal')
        return crossword
    result = find_word_vertical(crossword, word)
    if result:
        row_index, col_index = result
        capitalize_in_matrix(row_index, col_index, 'vertical')
        return crossword
    return crossword
crossword = [['s', 'd', 'o', 'g'],
             ['c', 'u', 'c', 'm'],
             ['a', 'c', 'a', 't'],
             ['t', 'e', 't', 'k']]
word = 'cat'
mod_crossword = capitalize_word_in_crossword(crossword, word)
print(mod_crossword)

````
